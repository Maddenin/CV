Let me explain how my ticket aggregator works.

### Logic blocks
First, we must understand the conditions of the task.
There are four criteria (the ticket price, the number of transfers, a refund, and luggage) and four offer categories (the best, good enough, bad, other) for flight classification. We must write a program that gets as input the data on these criteria. The data is entered from the keyboard. Then the program should classify an offer based on these criteria and print on the screen an offer category and this offer's description.

So, my program must:
* Be polite. Print welcome words for the user.
* Get input from the keyboard:
* As we have four criteria – there should be four categories: “the ticket price” must be a float, the number of transfers – integer, “refund” and ‘’ luggage’ – Boolean
* Transform the data and check whether the data are entered in a right way
* Classify an offer by four categories (the best, good enough, bad, other).
* Print on the screen an offer category and offer's description.

### Program blocks
I have used logic blocks structure for the program blocks, but expanded with technical moments:
* For the welcome text I use the function “welcome”. It’s very simple. The only moment to explain is "getpass" library to get the user’s login name. And I use variables for printing text separators.
* For keyboard input I use function "get_criteria". It expects keyboard input and returns four values (float, int, bool, bool). I use "try-input-except" construction for checking (by ValueError) whether the keyaboard input has the right format. Moreover, I use a separate function "get_bool" to match (by dictionary) keyboard input for TRUE \ FALSE Boolean states.
* For classification I use function classify. It has Boolean conditions and if-elif-else constructions.
* For result printing, I use the main function "result" and function "convert_bool" for Boolean states matching on people language, and class "text" for text formatting, and dictionary "offer_dict" for the offer's type and the offer's description matching.
* Al functionality (except welcome) is on while loop as long as the user wants the program to classify new offers.

```python
from tkinter import *
from tkinter import messagebox as msg
import getpass #for getting user login's name

root=Tk()
root.title('Ticket Aggregator')

def descr():
    msg.showinfo('Offer types description', '\'Best offer\': It\'s the one which \
costs less than $200; it is either a direct flight or a flight with one transfer; \
refund is included; luggage is included; pets allowed.\n\
\'Good enough offer\': It\'s price is in a range from $200 to $250; it is either \
a direct flight or might require one or two transfers; refund, luggage and pets \
is either included or not.\n\'Bad offer\': It\'s the one which price is more than \
$250 or three and more transfers are required; refund, luggage and pets is either \
included or not.\n\'Other offer\': It cannot be classified as either of three \
main categories.')

def classify():
    Price = price.get()
    Transfers = transfers.get()
    Refund = ref.get()
    Luggage = lug.get()
    Animal = anim.get()
    Offer_type = 'other offer'
    if Price < 200 and Transfers <= 1 and Refund and Luggage and Animal: Offer_type = 'the best offer'
    elif Price <= 250 and Transfers <= 2: Offer_type = 'good enough offer'
    elif Price > 250 or Transfers >= 3: Offer_type = 'the bad offer'
    msg.showinfo('Offer\'s type', f'I have classified this offer based on your \
criteria and decided that it is {Offer_type}')

price = Scale(root,orient=HORIZONTAL,length=800,from_=50,to=2000,tickinterval=150,
               resolution=1, label='The ticket price: ')
transfers = Scale(root,orient=HORIZONTAL,length=800,from_=0,to=5,tickinterval=1,
               resolution=1, label='The number of transfers: ')

lug = BooleanVar()
lug.set(0)
luggage = Checkbutton(text = "Luggage", variable = lug, onvalue = TRUE, offvalue = FALSE)
ref = BooleanVar()
ref.set(0)
refund = Checkbutton(text = "Refund", variable = ref, onvalue = TRUE, offvalue = FALSE)
anim = BooleanVar()
anim.set(0)
animal = Checkbutton(text = "Animal", variable = anim, onvalue = TRUE, offvalue = FALSE)
classify = Button(root,text="Classify offer ",relief = RAISED, command = classify)
description = Button(root,text="Offer types description",relief = RAISED, command = descr)

price.pack(anchor = W)
transfers.pack(anchor = W)
luggage.pack(anchor = W, side = LEFT)
refund.pack(anchor = W, side = LEFT)
animal.pack(anchor = W, side = LEFT)
classify.pack(side = RIGHT)
description.pack(side = RIGHT)

msg.showinfo('Wellcome', f'Hello {getpass.getuser()}, this is your ticket сlassification \
assistant.\nI will classify every flights from Moscow \
to New York and help to make the right decision.')
root.mainloop()
```
